```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as plticker
import seaborn as sns
import datetime

pd.options.mode.chained_assignment = None # supress some annoying warnings
```

```python
df = pd.read_csv("./data/opp-stops_state.csv")
```

```python
df.head()
```

```python
df_white = df.loc[df["subject_race"] == "white"]
```

```python
df_black = df.loc[df["subject_race"] == "black"]
```

```python
df_hisp = df.loc[df["subject_race"] == "hispanic"]
```

```python
df_black.reset_index(drop=True)
df_hisp.reset_index(drop=True)
df_white.reset_index(drop = True)
```

```python
white_hit_rate = df_white.dropna(subset=["search_rate"]).reset_index(drop = True)
```

```python
black_hit_rate = df_black.dropna(subset=["search_rate"]).reset_index(drop = True)
```

```python
hisp_hit_rate = df_hisp.dropna(subset=["search_rate"]).reset_index(drop = True)
```

## Thingy 

```python
def preprocessing(original_df_minor, df_white):
    reduced_df = original_df_minor[["subgeography", "search_rate", "stops_per_year", "geography"]]
    reduced_df["total_searches"] = reduced_df["search_rate"] * reduced_df["stops_per_year"] + df_white["search_rate"] * df_white["stops_per_year"]
    reduced_df.drop(["search_rate", "stops_per_year"], axis = 1, inplace=True)
    merged_df = original_df_minor.merge(reduced_df, on = ["subgeography", "geography"], suffixes=("", "_of_county"))
    return merged_df
```

```python
def plotting(df_white, merged_dfs, base, lim, metric, titles, label):
    fig, axs = plt.subplots(1, 2)
    fig.set_size_inches(16, 8)
    
    for i, merged_df in enumerate(merged_dfs):
        
        ax = axs[i]
        title = titles[i]
        
        ax.scatter(x = df_white[metric] * 100, y = merged_df[metric] * 100, s = merged_df["total_searches"], color="None", edgecolors = "black", linewidths=0.5)

        ax.set_xlim(0,lim)
        ax.set_ylim(0,lim)
        ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")
        loc = plticker.MultipleLocator(base=base) 
        ax.xaxis.set_major_locator(loc)
        ax.yaxis.set_major_locator(loc)
        ax.xaxis.set_tick_params(labelsize=20)
        ax.yaxis.set_tick_params(labelsize=20)
        
        if i == 0:
            ax.set_ylabel("Minority hit rate (%)", fontsize = 16)
        else:
            ax.set_yticks([])
        ax.set_title(f"{title} drivers", fontsize = 18)
        
        fig.text(0.5, 0, label, ha='center', fontsize = 18)
        ax.plot()
```

```python
merged_hisp = preprocessing(hisp_hit_rate, white_hit_rate)
merged_black = preprocessing(black_hit_rate, white_hit_rate)
plotting(white_hit_rate, [merged_black, merged_hisp], 20, 75, "hit_rate", ["Black", "Hispanic"], "White hit rate (%)")

merged_hisp = preprocessing(hisp_hit_rate, white_hit_rate)
merged_black = preprocessing(black_hit_rate, white_hit_rate)
plotting(white_hit_rate, [merged_black, merged_hisp], 10, 55, "inferred_threshold", ["Black", "Hispanic"], "White threshold (%)")
```
